```python

import matplotlib.pyplot as plt

# Scatter plot of Temperature vs Wind Speed
plt.figure(figsize=(10, 6))
plt.scatter(df['datetime'], df['temperature'], label='Temp vs datetime', color='tab:blue')
plt.title('Temperature vs datetime')
plt.xlabel('datetime ')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```
