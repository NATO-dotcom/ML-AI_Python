```python
%matplotlib inline
import matplotlib.pyplot as plt
from numpy.random import randn

# Generate 1000 random numbers
data = randn(1000)

# Plot histogram
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of randn(1000)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

CREATING MY DATASET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import randn
#creating a dataset
df=pd.DataFrame(randn(500,7),columns=['Mon','Tue','Wed','Thur','Fri','Sat','Sun'])
df
df.head()
```

HISTOGRAM

```python
df.hist(bins=30,figsize=(10,10))
plt.show()
```

```python
df.plot(kind='hist',figsize=(5,5),stacked=True,grid=True,bins=30)
```

```python
df.abs().plot(kind='hist',figsize=(5,5),stacked=True,grid=True,bins=30)
```

```python
df[['Thur','Wed','Sat']].abs().plot(kind='hist',figsize=(5,5),stacked=True,grid=True,bins=30)
```

AREA PLOT

```python
df.plot(kind='area',figsize=(10,5),stacked=False,grid=True,alpha=0.9)
```

```python
df.abs().plot.area(figsize=(10,5),stacked=False,grid=True,alpha=0.9,title="Area Plot")
plt.show()
```

```python
df['Mon'].abs().plot.area(figsize=(10,5),stacked=False,grid=True,alpha=0.9,title="Area Plot")
```

```python
df[['Mon','Fri']].abs().plot.area(figsize=(10,5),stacked=False,grid=True,alpha=0.9,title="Area Plot")
```

BAR CHART

```python
df.plot(kind='bar',figsize=(10,5),stacked=True,grid=True,alpha=0.9,title='Bar chart'),plt.show()
```

```python
df.abs().plot.bar(figsize=(10,5),stacked=True,grid=True,alpha=0.9,title='Bar chart',fontsize=10)
plt.show()
```

LINE CHART

```python
df.plot(kind='line',figsize=(10,5),grid=True,alpha=0.9,title='Line chart')
```

```python
df.abs().plot.line(figsize=(10,5),grid=True,alpha=0.9,title='Line chart',fontsize=10,lw=3),plt.show()
```

SCATTER PLOT

```python
df.plot(kind='scatter',x='Mon',y='Tue',figsize=(10,5),grid=True,alpha=0.9,title='Scatter plot')
```

```python
df.plot(kind='scatter',x='Mon',y='Tue',figsize=(10,5),grid=True,alpha=0.9,title='Scatter plot'),plt.show()
```

```python
df.plot(kind='scatter',x='Mon',y='Tue',figsize=(10,5),grid=True,alpha=0.9,title='Scatter plot',c=df['Wed']*100,cmap='cool')
```

```python
df.plot(kind='scatter',x='Mon',y='Tue',figsize=(10,5),grid=True,alpha=0.9,title='Scatter plot',c='Wed',cmap='berlin')
```

BOX PLOT

```python
df.plot(kind='box',figsize=(10,5),grid=True,title='Box plot')
```

```python
df.plot(kind='box',figsize=(10,5),title='Box plot',cmap='cool')
```

HEX BIN PLOT

```python
df.plot(kind='hexbin',x='Mon',y='Tue',figsize=(10,5),grid=True,title='Hex bin plot',gridsize=20)
```

```python
df.plot(kind='hexbin',x='Mon',y='Tue',figsize=(10,5),title='Hex bin plot',gridsize=20,cmap='magma')
```

KDE & DENSITY PLOTS

```python
df.plot(kind='density',figsize=(10,5),grid=True,title='Density plot')
```

```python
df.plot(kind='density',figsize=(10,5),grid=True,title='Density plot',lw=3)
```

```python
df.plot.kde(figsize=(10,5),title='KDE plot',lw=4,alpha=0.5)
```
