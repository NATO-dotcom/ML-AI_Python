```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('/content/Titanic-Dataset.csv')
data
```

LINEGRAPH

```python
plt.plot(data['Age'].dropna().sort_values().values)
plt.title('Age')
plt.title("sorted passenger ages")
plt.xlabel("Passenger Index")
plt.ylabel("Age")
plt.grid(True)#if false no grid
plt.show()
```

BARCHART

```python
survived_counts=data["Survived"].value_counts()
plt.bar(["Not Survived","Survived"],survived_counts,color=["red","green"])
plt.title("Survival Counts")
plt.ylabel("Number of passengers")
plt.xlabel("Counts")
plt.show()
```

```python
class_counts=data["Pclass"].value_counts()
plt.bar(["1st Class","2nd Class","3rd Class"],class_counts,color=["red","green","blue"])
plt.title("Class Counts")
plt.ylabel("Number of passengers")
plt.xlabel("Counts")
plt.show()
```

PIECHART

```python
classes=data["Pclass"].value_counts()
labels=["class1","class2","class3"]
color=["red","green","blue"]
plt.pie(classes,labels=labels,autopct="%1.2f%%")
plt.title("Passenger Class Distribution")
plt.show()
```

HISTOGRAM

```python
plt.hist(data["Age"].dropna(), bins=30, color="red", edgecolor="black")
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Number of passengers")
plt.show()
```

```python
plt.hist(data["Fare"].dropna(),bins=30,alpha=0.5,label="fare",color="red",edgecolor="black")
plt.hist(data["Age"].dropna(),bins=10,alpha=0.5,label="Age",color="blue",edgecolor="black")
plt.title("Fare and Age distribution")
plt.xlabel("value")
plt.ylabel("Number of passengers")
plt.legend()#gives the keys
plt.grid(True)
plt.show()
```

BOXPLOT

```python
plt.boxplot(data["Fare"].dropna())
plt.title("Fare Boxplot")
plt.ylabel("Fare")
plt.grid(True)
plt.show()
```

SCATTERPLOT

```python
plt.scatter(data["Age"],data["Fare"],alpha=0.5,c=data["Pclass"],cmap="plasma")
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.colorbar(label="Pclass")
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(12,6))
axs[0].hist(data["Age"].dropna(),bins=30,color="red",edgecolor="black")
axs[0].set_title("Age distribution Histogram")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Number of passengers")
axs[1].boxplot(data["Fare"].dropna())
axs[1].set_title("Fare distribution Boxplot")
plt.tight_layout()
plt.show()
```
