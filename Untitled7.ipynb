```python
%matplotlib inline
import matplotlib.pyplot as plt
from numpy.random import randn

# Generate 1000 random numbers
data = randn(1000)

# Plot histogram
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of randn(1000)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
